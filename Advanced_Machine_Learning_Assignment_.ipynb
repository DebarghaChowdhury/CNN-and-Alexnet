# Loading of Train Dataset

```python
files1=[]
labels1=[]
for i in os.listdir('train/'):
    for j in os.listdir('train/'+i):
        files1.append('train/'+i+'/'+j)
        labels1.append(i)
```

```python
PATH1 = 'train/'
classes1 = os.listdir(PATH1)
cl_path1=os.path.join(PATH1, classes1[1])
cl_data_path1=[os.path.join(cl_path1, img) for img in os.listdir(cl_path1) ]
len(cl_data_path1)
for i in range(0,5):
    img1 = mig.imread(cl_data_path1[i])
    print(i,img1.shape)
```

### Creation of Train Dataset

```python
train=pd.DataFrame()
```

```python
train['Files']=files1
train['Labels']=labels1
```

```python
train
```

```python
train.info
```

# Defining size of the image

```python
IMG_SIZE= 100
```

# Training Dataset

```python
import cv2
x_train=[]
y_train=[]
for t in range(len(train)):
    #folder = df1["Folder"][t]
    file1 = train["Files"][t]
    image = cv2.imread(file1,cv2.IMREAD_GRAYSCALE)
    if image is None:
        print('Null')
    else:
        image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
        image = image/255.0
        image = image.reshape(-1,IMG_SIZE,IMG_SIZE,1)
        image = image.reshape(image.shape[1], image.shape[2])
        if t % 500== 0:
            print(t)
        x_train.append(image)
        y_train.append(train["Labels"][t])
```

```python
x_train = np.array(x_train)
print("x_train done")
y_train = np.array(pd.get_dummies(y_train), dtype= np.uint8)
```

```python
x_train = x_train.reshape(-1,IMG_SIZE,IMG_SIZE, 1)
```

```python
x_train.shape
```

# Testing Dataset

```python
x_test=[]
y_test=[]
for t in range(len(test)):
    #folder = df2["Folder"][t]
    file1= test["Files"][t]
    image = cv2.imread(file1,cv2.IMREAD_GRAYSCALE)
    if image is None:
        print('Null')
    else:
        image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
        image = image/255.0
        image = image.reshape(1,IMG_SIZE,IMG_SIZE,1) 
        image = image.reshape(image.shape[1], image.shape[2])
        if t % 100 == 0:
            print(t)
        x_test.append(image)
        y_test.append(test["Labels"][t])
```

```python
x_test = np.array(x_test)
print("x_test done")
y_test = np.array(pd.get_dummies(y_test), dtype= np.uint8)
```

```python
x_test = x_test.reshape(-1,IMG_SIZE,IMG_SIZE, 1)
```

```python
x_test.shape
```

# Obtaining Libraries for CNN

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential, Model
from keras.layers import Dropout, Conv2D, Dense, BatchNormalization, MaxPooling2D, Flatten,Activation, Dense
```

# Implementation of Convolutional Neural Network (CNN)

```python
model = Sequential()
model.add(Conv2D(64, 3, 3, padding = 'same', input_shape=(x_train.shape[1:])))
model.add(Activation('relu'))
BatchNormalization(axis=1)
model.add(Conv2D(64,3,3 , padding = 'same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
BatchNormalization()
model.add(Dense(512))
model.add(Activation('relu'))
BatchNormalization()
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('softmax'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = "adam", metrics = ["accuracy"])
```

```python
model.summary()
```

```python
hist = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs = 100)
```

# Obtaining Accuracy Graph

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Obtaining Loss Graph

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Implementation of AlexNet

```python
model1=Sequential()
model1.add(Conv2D(filters=96,kernel_size=(11,11),strides=(4,4),padding="valid",activation="relu",input_shape=(100,100,1)))
model1.add(MaxPooling2D(pool_size=(1,1),strides=(2,2)))
model1.add(BatchNormalization())
model1.add(Conv2D(filters=256,kernel_size=(5,5),strides=(1,1),padding="valid",activation="relu"))
model1.add(MaxPooling2D(pool_size=(1,1),strides=(2,2)))
model1.add(BatchNormalization())
model1.add(Conv2D(filters=384,kernel_size=(3,3),strides=(1,1),padding="valid",activation="relu"))
model1.add(MaxPooling2D(pool_size=(1,1),strides=(2,2)))
model1.add(BatchNormalization())
model1.add(Flatten())
model1.add(Dense(4096,input_shape=(100,100,1),activation="relu"))
model1.add(Dropout(0.4))
model1.add(BatchNormalization())
model1.add(Dense(4096,activation="relu"))
model1.add(Dropout(0.4))
model1.add(BatchNormalization())
model1.add(Dense(1000,activation="relu"))
model1.add(Dropout(0.4))
model1.add(BatchNormalization())
model1.add(Dense(2,activation="softmax"))
model1.summary()
```

```python
model1.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model1.fit(x_train, y_train, epochs=20)
```
